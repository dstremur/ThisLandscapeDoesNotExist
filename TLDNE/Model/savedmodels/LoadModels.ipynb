```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt

```

```python
Model = tf.keras.models.load_model('Chollet200EpochsRMSProp128x128NoBatch100min.h5')
```

```python
Model.summary()
```

```python
anzahl = 1 
random_vector = tf.random.normal(shape=(anzahl, 128))
images = Model(random_vector)
images *=255

for x in range(anzahl):
    img = keras.utils.array_to_img(images[x])
    plt.figure()
    plt.axis("off")
    plt.imshow(img)
```
